### Diccionario de datos

- Código_ciudad: variable que codifica la ciudad del inmueble:
  - Mahattan = 1 
  - Bronx = 2
  - Brooklyn = 3
  - Queens = 4
  - Staten Island = 5
- Bloque y lote: combinados forman un indicador único empleado en NYC, conocido también como BBL
- Categoría actual y a la venta: describe el uso constructivo del inmueble. La segunda parte del código proporciona informacion adicional acerca del estilo de construccion:
  - A: vivienda unifamiliar. 
  - O: oficinas.
  - R: condominios.

```python
# Librerias que se utilizan 

import numpy as np # arrays
import pandas as pd # dataframes
import matplotlib.pyplot as plt # visualizaciones basicas
import seaborn as sns # visualizaciones algo mas elaboradas
```

```python
# NYC sales
nyc = pd.read_csv("./data/01_raw/nyc-rolling-sales.csv", encoding = "latin-1")
```

```python
nyc.columns = map(str.lower, nyc.columns) # a minusculas los nombres de las columnas 
# cambio el nombre de las columnnas con diccionario
```

```python
nyc
```

```python
nyc = nyc.rename(columns = {"unnamed: 0": "sin_nombre",
                     "borough": "codigo_ciudad",
                     "neighborhood": "barrio", 
                     "building class category": "categoria_inmueble",
                     "tax class at present": "tipo_impuesto_actual",
                     "block": "bloque",
                     "building class at present": "categoria_actual",
                     "adress": "direccion",
                     "apartament number": "numero",
                     "zip code": "codigo_postal",
                     "residential units": "unidades_residenciales",
                     "commercial units": "unidades_comerciales",
                     "total units": "total_unidades",
                     "land square feet": "terreno_metros_cuadrados",
                     "gross square feet": "superficie_total",
                     "year built": "fecha_construccion",
                     "tax class at time of sale": "impuesto_al_vender",
                     "building class at time of sale": "categoria_al_vender",
                     "sale price": "precio_venta",
                     "sale date": "fecha_venta"})
```
